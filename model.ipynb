```python
#Function to set valid column to 0 & add reason
def invalidate_column(name, reason):
    columns.loc[columns['name'] == name, 'valid'] = 0
    columns.loc[columns['name'] == name,'droped_reason'] = reason
    
#Function to set valid column to 1 & remove reason for being dropped
def revalidate_column(name):
    columns.loc[columns['name'] == name,'valid'] = 1
    columns.loc[columns['name'] == name, 'droped_reason'] = np.nan

#Function to print full dataframes
def print_full(x):
    pd.set_option('display.max_rows', len(x))
    print(x)
    pd.reset_option('display.max_rows')

    
invalidate_column('Id', 'Id has no logical correlation with povety')

print_full(columns)
```
